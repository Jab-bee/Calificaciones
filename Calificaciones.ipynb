```python
import pandas as pd

# Cargar el archivo
df = pd.read_csv("notas_estudiantes.csv")
```

```python
# Mostrar las 5 primeras filas
print("Primeras 5 filas del DataFrame:")
print(df.head())
```

```python
# Acceder a las columnas
asignatura = df['asignatura']
nota = df['nota']

# Agrupar por producto y sumar la columna 'Total'
resultado = df.groupby("asignatura")["nota"].mean()
print("Suma de 'Total' agrupada por 'Nota':")
print(resultado)
```

```python
import matplotlib.pyplot as plt
# Crear el gráfico de barras
resultado.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red'])

# Personalizar el gráfico
plt.title("Promedio de Notas por Asignatura")
plt.xlabel("Asignatura")
plt.ylabel("Nota Promedio")
plt.ylim(0, 10)  # Las notas están en una escala de 0 a 10
plt.grid(axis='y', linestyle='--')

# Mostrar el gráfico
plt.show()
```

```python
# Filtrar los datos para el estudiante específico
primer_estudiante = df[df["estudiante"] == "Estudiante_1"]

# Calcular el promedio de las notas por asignatura para el estudiante específico
resultado1 = primer_estudiante.groupby("asignatura")["nota"].mean()
print("Nota promedio por asignatura para Estudiante_1:")
print(resultado1)
```

```python
import matplotlib.pyplot as plt
# Crear el gráfico de barras
resultado1.plot(kind='bar', color=['black', 'yellow', 'darkblue', 'purple'])

# Personalizar el gráfico
plt.title("Promedio de Notas por Asignatura")
plt.xlabel("Asignatura")
plt.ylabel("Nota Promedio")
plt.ylim(0, 10)  # Las notas están en una escala de 0 a 10
plt.grid(axis='y', linestyle='--')

# Mostrar el gráfico
plt.show()
```
